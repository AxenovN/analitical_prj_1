```python
base_url = 'https://cloud-api.yandex.net/v1/disk/public/resources/download' 
public_key = 'https://disk.yandex.ru/d/kMrULQ9QCD7-CQ'

final_url = f'{base_url}?public_key={public_key}'
print(final_url)
```

```python
# type of getting file after .get()
```

```python
import requests

what_we_get = requests.get(final_url)
print(f'type of object: {type(what_we_get)}\n')
print(f'printing response: {what_we_get}\n')
print(f'dictionary: {what_we_get.json()}\n')
print(f"link itself: {what_we_get.json()['href']}")
```

```python
# look at file
```

```python
download_link = what_we_get.json()['href']
loading_response = requests.get(download_link)

print(f'Type of downloading response: {type(loading_response)}\n')
print(f'Content of file: \n{loading_response.content}')
```
